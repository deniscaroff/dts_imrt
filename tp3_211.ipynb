#UE 2.11 - TP3 : Interaction rayonnement-matière

```python
#Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.interpolate import CubicSpline
```

```python
#Données
data1 = pd.read_excel("tp3_photons_500keV.xlsx")
data2 = pd.read_excel("tp3_electrons_2MeV_dans_eau.xlsx")
data3 = pd.read_excel("tp3_protons_150MeV_dans_eau.xlsx")

print(data1.head())
print(data2.head())
print(data3.head())
```

```python
#Variables 1
Epaisseur1 = data1["Epaisseur_cm"]
Ratio_eau = data1["I/I0_Eau"]
Ratio_Al = data1["I/I0_Al"]
Ratio_Pb = data1["I/I0_Pb"]

#Variables 2
Epaisseur2 = data2["Epaisseur_cm"]
Energie2 = data2["Energie_restante_MeV"]

#Variables 3
Epaisseur3 = data3["Epaisseur_cm"]
Energie3 = data3["Energie_restante_MeV"]
TEL3 = data3["TEL_MeV/cm"]
```

```python
#Interpolation 1
Epaisseur1_interp = np.linspace(min(Epaisseur1), max(Epaisseur1), 200)
Ratio_eau_interp = CubicSpline(Epaisseur1, Ratio_eau)(Epaisseur1_interp)
Ratio_Al_interp = CubicSpline(Epaisseur1, Ratio_Al)(Epaisseur1_interp)
Ratio_Pb_interp = CubicSpline(Epaisseur1, Ratio_Pb)(Epaisseur1_interp)

#Interpolation 2
Epaisseur2_interp = np.linspace(min(Epaisseur2), max(Epaisseur2), 200)
Energie2_interp = CubicSpline(Epaisseur2, Energie2)(Epaisseur2_interp)

#Interpolation 3
Epaisseur3_interp = np.linspace(min(Epaisseur3), max(Epaisseur3), 200)
Energie3_interp = CubicSpline(Epaisseur3, Energie3)(Epaisseur3_interp)
TEL3_interp = CubicSpline(Epaisseur3, TEL3)(Epaisseur3_interp)
```

```python
#Graphe 1a
plt.figure(figsize=(10,6))
plt.scatter(Epaisseur1, Ratio_eau, marker='+', color='black', label='Matériau : A COMPLETER')
plt.plot(Epaisseur1_interp, Ratio_eau_interp, color='black')
plt.xlabel("A COMPLETER", fontsize=16)
plt.ylabel("A COMPLETER", fontsize=16)
plt.title("A COMPLETER", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graphe 1c
plt.figure(figsize=(10,6))
plt.scatter(Epaisseur1, Ratio_Pb, marker='+', color='black', label='Matériau : A COMPLETER')
plt.plot(Epaisseur1_interp, Ratio_Pb_interp, color='black')
plt.xlabel("A COMPLETER", fontsize=16)
plt.ylabel("A COMPLETER", fontsize=16)
plt.title("A COMPLETER", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
