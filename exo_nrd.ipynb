```python
#Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#Importation des données
data = pd.read_excel('exercice_nrd_python.xlsx')

print(data.head())
```

```python
#Variables
Age = data["Age"]
Taille = data["Taille"]
Masse = data["Masse"]
IMC = data["IMC"]
Ainj = data["Ainj"]
```

```python
#Corrélation âge-activité
r1, p_value1 = pearsonr(Age, Ainj)

print(f"Coefficient de corrélation de Pearson : r = {r1:.3f}")
print(f"p-value = {p_value1:.3f}")
```

```python
#Corrélation IMC-activité
r2, p_value2 = pearsonr(IMC, Ainj)

print(f"Coefficient de corrélation de Pearson : r = {r2:.3f}")
print(f"p-value = {p_value2:.3e}")
```

```python
#Corrélation Age-IMC
r3, p_value3 = pearsonr(Age, IMC)

print(f"Coefficient de corrélation de Pearson : r = {r3:.3f}")
print(f"p-value = {p_value3:.3f}")
```
