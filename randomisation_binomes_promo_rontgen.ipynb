#Randomisation des binômes de TP - DTS IMRT, Promotion Röntgen

```python
#Importation des bibliothèques nécessaires
import random

#Liste des étudiants
etudiants = ["Clément", "Lucas B.", "Florian", "Giresse", "Léa", "Enolane", "Camille", "Sabrina", "Maëlys", "Raphaëlla", "Lucas L.", "Lola", "Rose", "Faustine", "Maëlle", "Yoann", "Marceau", "Kilian", "Houari", "Miguel", "Nisrine", "Iliana", "Gamze", "Neyssa"]

#Randomisation
random.shuffle(etudiants)
```

```python
#Création des binômes
binomes = [(etudiants[i], etudiants[i+1]) for i in range (0, len(etudiants), 2)]

#Affichage
for i, (a, b) in enumerate(binomes, start=1):
  print(f"Binôme {i} : {a} et {b}")
```
