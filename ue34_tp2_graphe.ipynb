```python
#Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Importation des données
data = pd.read_excel("ue34_tp2_graphe.xlsx")

#Variables
Diam = data["Diam"]
Ratio = data["Ratio"]
CRC = data["CRC"]
```

```python
#Graphe Ratio
plt.figure(figsize=(10,6))
plt.plot(Diam, Ratio, marker='+', color='black', label='Valeurs expérimentales')
plt.axhline(y=10.137, color='red', linestyle='--', label='Ratio théorique')
plt.xlabel('Diamètre de la sphère (mm)', fontsize=14)
plt.ylabel('Ratio sphère / fond', fontsize=14)
plt.xlim(10, 38)
plt.ylim(0,10.5)
plt.title('Evolution du ratio sphère / fond avec le diamètre des sphères', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Graphe CRC
plt.figure(figsize=(10,6))
plt.plot(Diam, CRC, marker='+', color='black', label='Valeurs expérimentales')
plt.xlabel('Diamètre de la sphère (mm)', fontsize=14)
plt.ylabel('CRC', fontsize=14)
plt.xlim(10, 38)
plt.ylim(0,150)
plt.title('Evolution du CRC avec le diamètre des sphères', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
