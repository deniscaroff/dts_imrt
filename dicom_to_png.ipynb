```python
import os
os.environ["PYDICOM_USE_V2"] = "1"
from pathlib import Path
import pydicom
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from ipywidgets import interact
import imageio.v3 as iio
from pydicom.pixel_data_handlers import pylibjpeg_handler
pydicom.config.image_handlers = [pylibjpeg_handler]


input_folder = Path(r"C:\Users\denca\Documents\Physique-Chimie\DICOM")
output_folder = Path(r"C:\Users\denca\Documents\Physique-Chimie\DICOM_export")
output_folder.mkdir(exist_ok=True)
files = list(input_folder.iterdir())

slices = []

for f in files:
    ds = pydicom.dcmread(f)
    img = ds.pixel_array.astype(np.float32)
    instance = getattr(ds, "InstanceNumber", 0)
    slices.append((instance, img))

slices.sort(key=lambda x: x[0])
volume = np.stack([s[1] for s in slices], axis=0)

print("Volume shape :", volume.shape)
print("Nb slices :", volume.shape[0])
print("Image shape :", volume.shape[1], "x", volume.shape[2])

#Normalisation
vmin = np.percentile(volume, 1)
vmax = np.percentile(volume, 99)
volume_norm = np.clip((volume - vmin) / (vmax - vmin), 0, 1)
volume_16bit = (volume_norm * 65535).astype(np.uint16)
print("Exporte dans :", output_folder)

for i in range(volume_16bit.shape[0]):
    out_path = output_folder / f"slice_{i:03d}.png"
    iio.imwrite(out_path, volume_16bit[i])

print("Export terminé")
print("Nombre de slices :", volume_16bit.shape[0])

def show_slice(i):
    plt.figure(figsize=(5,5))
    plt.imshow(volume_norm[i], cmap='gray')
    plt.title(f"Slice {i}")
    plt.axis("off")
    plt.show()

interact(show_slice, i=(0, volume.shape[0]-1))
```
