```python
#Library import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.optimize import curve_fit
```

```python
#Propriétés N0 et t1/2
N0 = 100
T_s = 241200
T_min = T_s / 60
T_h = T_s /3600
T_j = T_s / 86400
T_an = T_s / (86400 * 365.25)

tau_s = T_s / np.log(2)
tau_min = T_min / np.log(2)
tau_h = T_h / np.log(2)
tau_j = T_j / np.log(2)
tau_an = T_an / np.log(2)
```

```python
def radioactive_decay(t, N0, tau):
  return N0 * np.exp(-t/tau)
```

```python
t_s = np.linspace(0, 8*T_s, 200)
t_min = np.linspace(0, 8*T_min, 200)
t_h = np.linspace(0, 8*T_h, 200)
t_j = np.linspace(0, 8*T_j, 200)
t_an = np.linspace(0, 8*T_an, 200)

N = radioactive_decay(t_h, N0, tau_h)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t_h, N,color='black', marker='+', label='Courbe de décroissance')
plt.xlabel("Temps t (heures)", fontsize=16)
plt.ylabel("Proportion de noyaux restants (%)", fontsize=16)
plt.title("Décroissance radioactive du molybdène-99", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
